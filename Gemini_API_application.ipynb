## Step 2 – Import Libraries

Import the required Python libraries.

````python
import pathlib, textwrap
import google.generativeai as genai
from IPython.display import display, Markdown

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda line: True))
````

````python
from google.colab import userdata
````

## Step 3 – Configure API Key

Configure the Gemini API securely.

````python
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

#we can also use like this
#GOOGLE_API_KEY ='Paste your API key here'
#genai.configure(api_key=GOOGLE_API_KEY)
````

````python
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
````

## Step 4 – Load Gemini Model

Initialize the Gemini Flash model for text generation.

````python
model = genai.GenerativeModel('gemini-2.5-flash')
````

## Step 5 – Generate AI response

Send a prompt to the Gemini model and get the response.

````python
%%time
response = model.generate_content("What is the meaning of Life?")
````

## Step 6 – Display Output

````python
to_markdown(response.text)
````

````python
response.prompt_feedback
````

````python
response.candidates
````

## Streaming Response

````python
%%time
response = model.generate_content("What is the meaning of Life?", stream=True)
````

````python
for chunk in response:
  print(chunk.text)
  print("_" * 80)
````

````python
response = model.generate_content("what is the meaning of Life?", stream=True)
````

````python
response.prompt_feedback
````

````python
try:
  response.text
except Exception as e:
  print(f'{type(e).__name__}:{e}')
````

##Image Understanding

````python
import PIL.Image

img = PIL.Image.open('image.jpg')
img
````

````python
model = genai.GenerativeModel('gemini-2.5-flash')
````

````python
response = model.generate_content(img)

to_markdown(response.text)
````

````python
response = model.generate_content(["write a short, engaging blog post on this picture. It should include a description of the nature in the photo and talk about my journey nature feeling.", img], stream=True)
response.resolve()
````

````python
to_markdown(response.text)
````

##Chat History

````python
model = genai.GenerativeModel('gemini-2.5-flash')
chat = model.start_chat(history=[])
chat
````

````python
response = chat.send_message("In one sentence, explain how a computer works to a young child.")
to_markdown(response.text)
````

````python
chat.history
````

````python
response = chat.send_message("Okay, how about a more detailed explanation to a high schooler?", stream=True)

for chunk in response:
  print(chunk.text)
  print("_"*80)
````

````python
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
````
